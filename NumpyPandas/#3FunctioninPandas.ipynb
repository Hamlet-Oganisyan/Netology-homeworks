```python
import pandas as pd
movies=pd.read_csv('C:/Netology/Homeworksfiles/ml-latest-small/ratings.csv')
movies.head()
```

```python
def class_(x):
    """Классификация рейтингов фильмов"""
   
    if x <= 2:
        return ('низкий рейтинг')
    elif x <= 4:
        return ('средний рейтинг')
    else: 
        return ('высокий рейтинг')
        
movies['class'] = movies['rating'].apply(class_)
movies
```

## Задание 2

```python
keyword=pd.read_csv('C:/Netology/Homeworksfiles/ml-latest-small/keywords.csv')
keyword
```

```python
def region(row):
    """Гео-классификатор на запросы"""
    if 'москва'in row['keyword'] or 'тула' in row['keyword'] or 'ярославль' in row['keyword']:
        return 'Центр'
    elif 'петербург' in row['keyword'] or 'псков' in row['keyword'] or 'псков' in row['keyword']:
        return 'Северо-Запад'
    elif 'владивосток' in row['keyword'] or 'сахалин' in row['keyword'] or 'хабаровск' in row['keyword']:
        return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
keyword['region'] = keyword.apply(region, axis=1)
keyword
```

## Задание 3

```python
movies=pd.read_csv('C:/Netology/Homeworksfiles/ml-latest-small/movies.csv')
movies
```

```python
ratings=pd.read_csv('C:/Netology/Homeworksfiles/ml-latest-small/ratings.csv')
ratings
```

```python
new_df = movies.merge(ratings, on=['movieId'])
new_df
```

```python
years = []
for y in range(1950, 2011, 1):
    years.append(y)
print(years)
```

```python
def production_year(title):
    for y in years:
        if str(y) in title:
            return y
      
    return 1900
```

```python
new_df['production_year'] = new_df['title'].apply(production_year)
new_df
```

```python
mean_rating_year = new_df.groupby('production_year').rating.mean().reset_index()
mean_rating_year.sort_values('rating', ascending = False)
```
