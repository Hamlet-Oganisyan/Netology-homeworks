## Задание 3

```python
rating_df = pd.read_csv('C:/Netology/Homeworksfiles/ml-latest-small/ratings.csv')
rating_df.head()
```

```python
#Сгруппируем по UserID и посчитаем количество оценок каждого из них
id_rating_count = rating_df.groupby('userId').rating.count().reset_index()
id_rating_count.rename(columns = {'rating': 'rating_count'}, inplace=True)
id_rating_count
```

```python
#Сгруппируем по UserID поставивших более 100 оценок
Id100 = id_rating_count.loc[id_rating_count.rating_count > 100]
Id100
```

```python
#Отфильтруем основной датафрейм по UserID поставивших более 100 оценок

rating_df_100 = rating_df[rating_df['userId'].isin(Id100['userId'])]
rating_df_100.head()
```

```python
#Определим для каждого UserID min и max значения timestamp
timestamp = rating_df_100.groupby('userId').agg(['min', 'max'])['timestamp']
timestamp.head()
```

```python
#Посчитаем среднее время жизни пользователей, поставивших более 100 оценок путем  разницы между
#максимальным и минимальным значением столбца timestamp
def mean_(row):
    return timestamp['max'] - timestamp['min']
        
timestamp['mean'] = timestamp.apply(mean_)
timestamp
```
