## Задание 2

```python
data=pd.read_csv('C:/Netology/Homeworksfiles/power.csv')
data.head()
```

```python
pribaltic = data[(data['country'] == 'Lithuania')|(data['country'] =='Latvia')|(data['country'] =='Estonia')]
pribaltic.head(100)
```

```python
years_quantity = pribaltic[(pribaltic['year']>=2005)|(pribaltic['year']<=2010)|(pribaltic['quantity']>0)]
years_quantity.head()
```

```python
category = years_quantity[years_quantity.category.isin([4,12,21])]
category.head(100)
```

```python
summa = category['quantity'].sum()
print('Суммарное потребление: ', summa)
```
