## Урок "Визуализация данных"

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 15, 7
```

```python
# Мы берем данные об уровне счастья по миру за 2019 г. 
hapiness_data = pd.read_csv('C:/Netology/Homeworksfiles/World Happiness Report_2019.csv')
hapiness_data.head()
```

```python
hapiness_data.info()
```

```python
#Узнаем количество пустых строк в данных
hapiness_data.isnull().sum()
```

```python
# Построим визуализацию взаимосвязи уровня счастья страны от уровня социальной поддержки в этих странах
# Хотя, на показатель уровеня счастья влияют много факторов, но все же, мы видим на графике, что в большинстве своем
# с увеличением соцподдержки увеличивается и осознание счастья. Таким образом, можно сделать вывод, что есть некая 
# взамосвязь между уровнем социальной поддержки государства населения и осознания ими счастья.

plt.plot(hapiness_data['Score'], hapiness_data['Social support'])
plt.title('Динамика уровня счастья стран взависимости от социальной поддержки')
plt.xlabel('Уровень счастья')
plt.ylabel('Уровень социальной поддержки')
```

```python
# Также можем отразить визуализацию по всем показателям сразу.
# Намой взгляд это дает нам увидеть вцелом какие показатели важны для "осознанности счастья"
hapiness_data.plot(x = 'Score', y =['GDP per capita', 'Social support', 'Healthy life expectancy', \
                             'Freedom to make life choices', 'Generosity','Perceptions of corruption'],
                                          
                             title='Динамика уровня счастья стран взависимости от социальных параметров', 
                             xlabel='Уровень счастья', 
                             ylabel='Уровени социальных параметров')
```

```python
# На полученной визуализации можно увидеть, что такие показатели, как 'Generosity' и 'Perceptions of corruption 
# не имеют тенденции к росту с увеличением показателя уровня счастья. Можно предположить, что данные показатели не 
# влияют на увелечиния уровеня частья.
# Такой показатель, как 'Freedom to make life choices', хоть и имеет некоторый восходящий тренд с увеличением 
# уровнем счастья, но все же не очень сильний. 
# Но, что можно отчетливо увидеть, это то, в показателях 'GDP per capita', 'Social support', 'Healthy life expectancy', 
# при их увелечении, увеличивается и уровень счастья.
# Можно сделать предположение, что данные материальные категории непосредственно влияют на уровень счастья.
# Также визуально можно увидеть, что на всех приведенных графиках , а особенно у 'GDP per capita', 'Social support', 
# 'Healthy life expectancy', 'Freedom to make life choices' начиная от уровня счастья  6,4 начинается стабильный 
# восходящий тренд. Из этого я делаю вывод, что в странах с данным уровнем счатья, эти показатели очень важны для 
# осознания счастья.
# Сделаем выборку по эти показателям, чтобы увидеть что это за страны.

df = hapiness_data.loc[hapiness_data['Score'] >= 6.4]
df
```

```python
# Как я и предполагал, в оcновном это страны Америки, Европы и другие развитые страны.
# Последние Чили и Гватемалла, все же, наверное можно считать выбросами в данном списке.
# Давайте создадим точечную диаграмму и посмотрим на взаимосвязь между уровнем счастья и продолжительностью жизни 
# в этих странах

df.plot(kind= 'scatter', x = 'Score',  y = 'Healthy life expectancy',
                             title= 'Влияние продолжительности жизни на уровень счастья',
                             xlabel= 'Уровень счастья',
                             ylabel= 'Уровень продолжительности жизни')
```

```python
# На диаграмме видно, что для стран с отметкой уровня счастья 6,85 и выше данный показатель является относительно значимым
```

```python
# Вообще, из нашего списка видно, что в ТОПе "счастливых" стран Скандинавские или страны Северной Европы 
# И мне стало интересно, а как обстоит дело с "счастьем" в странах бавшего СССР.
# Создадим визууализацию по постсоветским странам.
```

```python
Ussr_hapiness_data = hapiness_data[(hapiness_data['Country or region'] == 'Russia')| \
                                   (hapiness_data['Country or region'] == 'Ukraine')| \
                                   (hapiness_data['Country or region'] == 'Belarus') | \
                                   (hapiness_data['Country or region'] == 'Uzbekistan')| \
                                   (hapiness_data['Country or region'] == 'Kazakhstan')| \
                                   (hapiness_data['Country or region'] == 'Turkmenistan')|\
                                   (hapiness_data['Country or region'] == 'Azerbaijan')|
                                   (hapiness_data['Country or region'] == 'Tajikistan') | \
                                   (hapiness_data['Country or region'] == 'Kyrgyzstan') | \
                                   (hapiness_data['Country or region'] == 'Georgia') | \
                                   (hapiness_data['Country or region'] == 'Moldavia') | \
                                   (hapiness_data['Country or region'] == 'Armenia') | \
                                   (hapiness_data['Country or region'] == 'Lithuania') | \
                                   (hapiness_data['Country or region'] == 'Latvia') | \
                                   (hapiness_data['Country or region'] == 'Estonia')]
Ussr_hapiness_data                                  
```

```python
Ussr_hapiness_data.plot(kind= 'barh', x = 'Country or region',  y = 'Score',
                             title= 'Кто самый счастливый на постсоветском пространстве',
                             xlabel= 'Страны',
                             ylabel= 'Уровень счастья')
```

```python
# Из приведенной визуализации мы видим, что на постсоветском пространстве самой "счастливой" страной в 2019 г. был
# Узбекистан, а самой "несчастливой" Украина.
```
