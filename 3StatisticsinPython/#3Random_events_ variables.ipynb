```python
import numpy as np
import random
import time


# Сказать честно, не совсем понятна задача, но как понял реализую ))))
# Имеется перечень выигрышных комбинаций лотерейного автомата:
# 777: 200 руб.
# 999: 100 руб.
# 555: 50 руб.
# 333: 15 руб.
# 111: 10 руб.
# *77: 5 руб.
# **7: 3 руб.
# *00: 2 руб.
# **0: 1 руб.

# Определим возможное количество всех выигрышных комбинаций при выпадании чисел от 000 до 999
gain_combination = 0
for i in range(0, 1000):
    i = '{:03}'.format(i)
    win_numbers = ['999', '777', '555', '333', '111']   
    if i in win_numbers or i.endswith('0') or i.endswith('00') or i.endswith('7') or i.endswith('77'):
        gain_combination += 1
        print(i, end = " ")

print('')
print('-'*50)      
print('Количество выигрышных комбинаций', gain_combination)  
```

```python
# Таким образом у нас 204 возможных выигрышных комбинаций.
# Определим теоритическую вероятность выигрыша "Р"
# где gain_combination - это количество возможных выигрышных комбинаций, а 1000 - количество возможных выпадаемых чисел
p = gain_combination/1000
p
# Теоритическая вероятность выпада любого выиграшного числа на лотерейном автомате составляет 1/5
```

```python
# Отвечаем на вопросы:
# - Выгодна ли игра игроку?
# - Сколько в среднем приобретает или теряет игрок за одну игру?

# Что подразумевается под фразой "за одну игру" ? Если то, что он 1 раз заплатил 1 руб. и попытал счастье, 
# но ему не повезло,то значит он потреяет всего 1 руб.
# А приобрести он может от 1 до 200 руб. И то, если выиграл 1 руб., то в ноль сработал, никакой прибыли )))
# Но, чтобы посчитать средний возможный выигрыш сумируем все возможные выигрыши и поделим на количество выигрышных чисел
gain = 0
gain_0 = 0
gain_00 = 0
gain_7 = 0
gain_77 = 0
gain_111 = 0
gain_333 = 0
gain_555 = 0
gain_777 = 0
gain_999 = 0
gain_combination = 204

for i in range(0, 1000):
    i = '{:03}'.format(i)
    if i.endswith('00'):
        gain_00 += 1
        print(i, end = " ")
    elif i.endswith('0'):
        gain_0 += 1
        print(i, end = " ")
    elif i == '777':
        gain_777 = 1
        print(i, end = " ")
    elif i.endswith('77'):
        gain_77 += 1
        print(i, end = " ")
    elif i.endswith('7'):
        gain_7 += 1
        print(i, end = " ")
    elif i == '111':  
        gain_111 += 1
        print(i, end = " ")
    elif i == '333':
        gain_333 += 1
        print(i, end = " ")
    elif i == '555':
        gain_555 += 1
        print(i, end = " ")   
    elif i == '999':
        gain_999 += 1
        print(i, end = " ") 
    
gain_total = gain_0*1+gain_00*2+gain_7*3+gain_77*5+gain_111*10+gain_333*15+gain_555*50+gain_999*100+gain_777*200

print('')
print('-'*50)      
print('Общая сумма всех выигрышей', gain_total, 'руб.')
print('Кол-во возможных выигрышей по 1 руб.', gain_0)
print('Кол-во возможных выигрышей по 2 руб.', gain_00)
print('Кол-во возможных выигрышей по 3 руб.', gain_7)
print('Кол-во возможных выигрышей по 5 руб.', gain_77)
print('Кол-во возможных выигрышей по 10 руб.', gain_111)
print('Кол-во возможных выигрышей по 15 руб.', gain_333)
print('Кол-во возможных выигрышей по 50 руб.', gain_555)
print('Кол-во возможных выигрышей по 100 руб.', gain_999)
print('Кол-во возможных выигрышей по 200 руб.', gain_777)
print('Количество выигрышных комбинаций', gain_combination)  
print('Средний выигрыш', gain_total/gain_combination, 'руб.')  
```
